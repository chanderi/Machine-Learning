# Building Model

```python
X=df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Spliting the data into Training Data(70%)and Testing Data(30%).
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
# A standard scaler converts a distribution, such that it has 0 mean and 1 standard deviation. 
# Standard scaler is used to convert all values in same magnitude
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()

X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

# Models and finding out the Best one

# Logistic Regression

```python
# Here we try to fit in different models and choose best one in them.
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
lr=LogisticRegression()
model1=lr.fit(X_train,y_train)
prediction1=model1.predict(X_test)
```

```python
# A confusion matrix is a table that is often used to describe the performance of a classification
# model (or “classifier”) on a set of test data for which the true values are known. 
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,prediction1)
cm

# tp fp
# fn tn
# Here 110 are the correctly identified cases - truely malignant cases
#       5 are the incorrectly identified cases - falsely malignant cases
#       2 are the incorrectly rejected - falsely benign cases
#      54  are the correctly rejected - truely benign cases
```

```python
plt.figure(figsize=(6,6))
c_mat = metrics.confusion_matrix(prediction1, y_test)
sns.heatmap(cm,annot=True,fmt='d',cbar=False)
plt.xlabel("Predictions",fontsize=10)
plt.ylabel("Truth",fontsize=10)
plt.title("Logistic Regression Confusion Matrics",fontsize=16)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction1)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,threshold = roc_curve(y_test,prediction1)
print("ROC_AUC score: ",roc_auc_score(y_test,prediction1))
plt.figure(figsize=(10,7))
plt.plot(fpr,tpr,"^-y",markerfacecolor='darkred',label="=LogisticRegressionClassifier")
plt.plot([0,1],[0,1],linestyle="--",color="red")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

# K Nearest Neighbor (KNN)     
# Support Vector Machine     
# Naive Bayes

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
models=[]

models.append(('KNN', KNeighborsClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
# KFold classifier is a resampling technique which evalutes the machine learning algorithm.
# It divides data set into k-parts.
# cross_val_score is used to reshuful the data.
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# evaluate each model

results =[]
names=[]
for name , model in models:
    kfold=KFold(n_splits=10, random_state=40)
    cv_results= cross_val_score(model, X_train, y_train, cv=kfold,scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    
    msg= '%s: cross_val_score:%f, Standard Deviation:(%f)' % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# make predictions on test datasets

SVM = SVC()
SVM.fit(X_train, y_train)
predictions= SVM.predict(X_test)
print(accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

# We are getting best accuracy with SVM which is 96.49%,the model is predicting with 96% accuracy on our test Data.
